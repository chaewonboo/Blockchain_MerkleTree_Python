```python
machine_index = np.array(File_GPU_1[:, 1])
machine_index = np.array(machine_index, dtype = int)
power_used_per = np.array(File_GPU_1[:, 3])
power_used_per = np.asfarray(power_used_per, float)
```

```python
print(machine_index)
print(power_used_per)
print(power_used_per[1])
```

```python
temp =[]
machine_1_array = np.array(temp, dtype = np.float32)
machine_2_array = np.array(temp, dtype = np.float32)
machine_3_array = np.array(temp, dtype = np.float32)
machine_4_array = np.array(temp, dtype = np.float32)
```

```python
for index in range(len(machine_index)):
    if machine_index[index] == 0:
        machine_1_array = np.append(machine_1_array, power_used_per[index])
    elif machine_index[index] == 1:
        machine_2_array = np.append(machine_2_array, power_used_per[index])
    elif machine_index[index] == 2:
        machine_3_array = np.append(machine_3_array, power_used_per[index])
    else:
        machine_4_array = np.append(machine_4_array, power_used_per[index])
```

```python
print(machine_1_array)
print(machine_2_array)
print(machine_3_array)
print(machine_4_array)
```

```python
print(len(machine_1_array))
```

```python
x= list(range(45))
plt.plot(x, machine_1_array, label='Trainer_1')
plt.plot(x, machine_2_array, label='Trainer_2')
plt.plot(x, machine_3_array, label='Trainer_3')
plt.plot(x, machine_4_array, label='Trainer_4')

plt.legend()

plt.xlabel("Machine_Learning_Time")
plt.ylabel("Power_Used_Per")
plt.title("Trainer_Used_Power")
plt.xlim([0, 47])
plt.ylim([35, 55])
plt.show()
```
